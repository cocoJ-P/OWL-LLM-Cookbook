ç”±äºèšåˆæœ¬ä½“ç­–ç•¥éœ€è¦æœ¬ä½“çš„URIè¿›è¡Œé“¾æ¥ï¼Œæ‰€ä»¥åœ¨å•æ–‡ä»¶çš„åŠ è½½ä¹‹å¤–ï¼Œæˆ‘ä»¬è¿˜éœ€è¦è¿›è¡ŒURIä¸æœ¬ä½“åœ°å€çš„æ˜ å°„ã€‚

ï¼ˆä¸ºäº†é™ä½ç®—åŠ›è´Ÿæ‹…ï¼Œæˆ‘ä»¬é‡‡ç”¨æœ¬åœ°çš„ç®€åŒ–ç‰ˆæœ¬MechOn-Fluidè¿›è¡Œæµ‹è¯•ï¼Œæ‰€ä»¥æˆ‘ä»¬çš„æ˜ å°„æ–‡ä»¶å…¨éƒ¨ä½¿ç”¨æœ¬åœ°åœ°å€ï¼‰

````python
from rdflib import Graph

AggregatedUrl = "./examples/2/mechon-fluid.owl"

# å®šä¹‰ URI â†’ æœ¬åœ°æ–‡ä»¶è·¯å¾„çš„æ˜ å°„å…³ç³»ï¼ˆæ¨¡æ‹Ÿ catalog.xml çš„åŠŸèƒ½ï¼‰
uri_map = {
    "https://purl.mechon.org/ont/calc/0.1.0": "./examples/2/calc.core.ttl",
    "https://purl.mechon.org/ont/mat/0.1.0": "./examples/2/mat.core.ttl",
    "https://purl.mechon.org/ont/geo/0.1.0": "./examples/2/geo.core.ttl",
}
# åŠ è½½ä¸»å›¾
g = Graph().parse(AggregatedUrl, format="turtle")

# éå† owl:imports å¹¶æ‰‹åŠ¨åŠ è½½æ˜ å°„æ–‡ä»¶
for s, p, o in g.triples((None, g.namespace_manager.expand_curie("owl:imports"), None)):
    uri = str(o)
    if uri in uri_map:
        g.parse(uri_map[uri], format="turtle")
    else:
        print(f"âš ï¸ æœªçŸ¥ import URI: {uri}ï¼Œè·³è¿‡æˆ–è€ƒè™‘ä¸‹è½½")
````

è®©æˆ‘ä»¬æ¥æ£€æŸ¥æ˜¯å¦æˆåŠŸè½½å…¥èšåˆæ–‡ä»¶

````python
import subprocess, tempfile, os
from rdflib import Graph, Namespace
from rdflib.namespace import RDF, RDFS, OWL, URIRef

def check_imports():
    """æ£€æŸ¥mechon-fluid.owlæ–‡ä»¶æ˜¯å¦æˆåŠŸå¯¼å…¥äº†æ‰€æœ‰èšåˆçš„å†…å®¹"""
    
    print("=== æ£€æŸ¥ mechon-fluid.owl å¯¼å…¥æƒ…å†µ ===\n")
    
    # 1. æ£€æŸ¥å£°æ˜çš„å¯¼å…¥
    print("1. å£°æ˜çš„å¯¼å…¥ (owl:imports):")
    ontology_uri = URIRef("https://purl.mechon.org/ont/fluid")
    for s, p, o in g.triples((ontology_uri, OWL.imports, None)):
        print(f"   - {o}")
    
    print("\n2. æ£€æŸ¥å„ä¸ªæ ¸å¿ƒæ–‡ä»¶çš„å†…å®¹æ˜¯å¦è¢«åŒ…å«:")
    
    # 2. æ£€æŸ¥å„ä¸ªæ ¸å¿ƒæ–‡ä»¶çš„å†…å®¹
    core_files = {
        "calc": "./examples/2/calc.core.ttl",
        "geo": "./examples/2/geo.core.ttl", 
        "mat": "./examples/2/mat.core.ttl",
    }
    
    for name, file_path in core_files.items():
        print(f"\n   --- {name.upper()} æ ¸å¿ƒæ–‡ä»¶æ£€æŸ¥ ---")
        
        # åŠ è½½æ ¸å¿ƒæ–‡ä»¶
        core_g = Graph().parse(file_path, format="turtle")
        
        # åªç»Ÿè®¡ URIRef ç±»å‹çš„å‘½åå®ä½“
        core_classes = [s for s in core_g.subjects(RDF.type, OWL.Class) if isinstance(s, URIRef)]
        core_obj_props = [s for s in core_g.subjects(RDF.type, OWL.ObjectProperty) if isinstance(s, URIRef)]
        core_data_props = [s for s in core_g.subjects(RDF.type, OWL.DatatypeProperty) if isinstance(s, URIRef)]
        
        print(f"   æ ¸å¿ƒæ–‡ä»¶åŒ…å«:")
        print(f"   - ç±»: {len(core_classes)} ä¸ª")
        # for cls in core_classes:
        #     print(f"     Â· {cls}")
        print(f"   - å¯¹è±¡å±æ€§: {len(core_obj_props)} ä¸ª")
        # for prop in core_obj_props:
        #     print(f"     Â· {prop}")
        print(f"   - æ•°æ®å±æ€§: {len(core_data_props)} ä¸ª")
        # for prop in core_data_props:
        #     print(f"     Â· {prop}")
        
        # æ£€æŸ¥ä¸»æ–‡ä»¶ä¸­æ˜¯å¦åŒ…å«è¿™äº›å†…å®¹
        found_classes = 0
        found_obj_props = 0
        found_data_props = 0
        
        for cls in core_classes:
            if (cls, RDF.type, OWL.Class) in g:
                found_classes += 1
        
        for prop in core_obj_props:
            if (prop, RDF.type, OWL.ObjectProperty) in g:
                found_obj_props += 1
                
        for prop in core_data_props:
            if (prop, RDF.type, OWL.DatatypeProperty) in g:
                found_data_props += 1
        
        print(f"   ä¸»æ–‡ä»¶ä¸­æ‰¾åˆ°:")
        print(f"   - ç±»: {found_classes}/{len(core_classes)} ä¸ª")
        print(f"   - å¯¹è±¡å±æ€§: {found_obj_props}/{len(core_obj_props)} ä¸ª")
        print(f"   - æ•°æ®å±æ€§: {found_data_props}/{len(core_data_props)} ä¸ª")
        
        # è®¡ç®—å¯¼å…¥æ¯”ä¾‹
        if len(core_classes) > 0:
            class_ratio = found_classes / len(core_classes) * 100
            print(f"   - ç±»å¯¼å…¥æ¯”ä¾‹: {class_ratio:.1f}%")
        
        if len(core_obj_props) > 0:
            obj_prop_ratio = found_obj_props / len(core_obj_props) * 100
            print(f"   - å¯¹è±¡å±æ€§å¯¼å…¥æ¯”ä¾‹: {obj_prop_ratio:.1f}%")
            
        if len(core_data_props) > 0:
            data_prop_ratio = found_data_props / len(core_data_props) * 100
            print(f"   - æ•°æ®å±æ€§å¯¼å…¥æ¯”ä¾‹: {data_prop_ratio:.1f}%")
    
    # 3. æ£€æŸ¥ä¸»æ–‡ä»¶çš„æ€»ç»Ÿè®¡
    print(f"\n3. ä¸»æ–‡ä»¶ (mechon-fluid.owl) æ€»ç»Ÿè®¡:")
    main_classes = list(g.subjects(RDF.type, OWL.Class))
    main_obj_props = list(g.subjects(RDF.type, OWL.ObjectProperty))
    main_data_props = list(g.subjects(RDF.type, OWL.DatatypeProperty))
    
    print(f"   - æ€»ç±»æ•°: {len(main_classes)}")
    print(f"   - æ€»å¯¹è±¡å±æ€§æ•°: {len(main_obj_props)}")
    print(f"   - æ€»æ•°æ®å±æ€§æ•°: {len(main_data_props)}")
    
    # 4. æ£€æŸ¥æ˜¯å¦æœ‰QUDTç›¸å…³çš„å¯¼å…¥
    print(f"\n4. QUDTç›¸å…³æ£€æŸ¥:")
    qudt_entities = [s for s, p, o in g.triples((None, None, None)) 
                    if "qudt.org" in str(s) or "qudt.org" in str(o)]
    print(f"   - QUDTç›¸å…³å®ä½“æ•°: {len(qudt_entities)}")
    
    # 5. æ£€æŸ¥å‘½åç©ºé—´
    print(f"\n5. å‘½åç©ºé—´æ£€æŸ¥:")
    namespaces = g.namespaces()
    for prefix, uri in namespaces:
        print(f"   - {prefix}: {uri}")

if __name__ == "__main__":
    check_imports()
````

````python
from rdflib import Graph, OWL, RDF
from rdflib.namespace import RDFS, Namespace
from collections import Counter
import textwrap
import tempfile

AggregatedUrl = "./examples/2/mechon-fluid-bridge.owl"

# å®šä¹‰ URI â†’ æœ¬åœ°æ–‡ä»¶è·¯å¾„çš„æ˜ å°„å…³ç³»ï¼ˆæ¨¡æ‹Ÿ catalog.xml çš„åŠŸèƒ½ï¼‰
uri_map = {
    "https://purl.mechon.org/ont/calc/0.1.0": "./examples/2/calc.core.ttl",
    "https://purl.mechon.org/ont/mat/0.1.0": "./examples/2/mat.core.ttl",
    "https://purl.mechon.org/ont/phy/0.1.0": "./examples/2/phy.core.ttl",
    "https://purl.mechon.org/ont/geo/0.1.0": "./examples/2/geo.core.ttl",
    "https://purl.mechon.org/ont/bridge/qk-alignment/0.1.0": "./examples/2/qk-alignment.ttl",
    "https://purl.mechon.org/ont/bridge/qk-bridge/0.1.0": "./examples/2/qudt-clip/qk-bridge.ttl",
}

def load_with_imports(graph, path_or_uri, uri_map, loaded=set(), loaded_files=[]):
    """é€’å½’åŠ è½½æŸæœ¬ä½“åŠå…¶æ‰€æœ‰ owl:importsï¼›è¿”å›å·²åŠ è½½æ–‡ä»¶åˆ—è¡¨"""
    graph.parse(path_or_uri, format="turtle")
    loaded_files.append(path_or_uri)

    for _, _, o in graph.triples((None, OWL.imports, None)):
        uri = str(o)
        if uri in loaded:
            continue
        if uri in uri_map:
            print(f"ğŸ”  å¯¼å…¥ {uri}")
            loaded.add(uri)
            load_with_imports(graph, uri_map[uri], uri_map, loaded, loaded_files)
        else:
            print(f"âš ï¸  æœªçŸ¥ import URI: {uri}ï¼ˆæœªåŠ è½½ï¼‰")
    return loaded_files
    
def summarize_graph(graph):
    """è¿”å›ä¸€ä¸ªå­—å…¸ï¼ŒåŒ…å«ä¸»è¦ç»Ÿè®¡ä¿¡æ¯"""
    summary = {}

    # -------- åŸºæœ¬è®¡æ•° --------
    summary["Triples"] = len(graph)
    summary["Classes"] = len(set(graph.subjects(RDF.type, OWL.Class)))
    summary["ObjectProperties"] = len(set(graph.subjects(RDF.type, OWL.ObjectProperty)))
    summary["DatatypeProperties"] = len(set(graph.subjects(RDF.type, OWL.DatatypeProperty)))
    summary["AnnotationProperties"] = len(set(graph.subjects(RDF.type, OWL.AnnotationProperty)))

    # ä»¥â€œå®ä¾‹â€ç²—ç•¥å®šä¹‰ä¸ºï¼šrdf:type ä¸ç­‰äº *Property / Class / Ontology
    system_types = {
        OWL.Class,
        OWL.ObjectProperty,
        OWL.DatatypeProperty,
        OWL.AnnotationProperty,
        OWL.Ontology,
    }
    individuals = {
        s for s, o in graph.subject_objects(RDF.type) if o not in system_types
    }
    summary["Individuals"] = len(individuals)

    # -------- å‘½åç©ºé—´åˆ†å¸ƒ --------
    # ç»Ÿè®¡æ¯ä¸ªå‘½åç©ºé—´å‰ç¼€å‡ºç°çš„ä¸‰å…ƒç»„æ•°é‡ï¼ˆæŒ‰ subject URI åˆ¤å®šï¼‰
    ns_counter = Counter()
    for s, _, _ in graph:
        if isinstance(s, Namespace):  # ç¨€æœ‰æƒ…å†µ
            continue
        if s.startswith("http"):
            ns = "/".join(s.split("/")[:-1])  # å–åˆ°å€’æ•°ç¬¬ä¸€çº§å‰çš„ç‰‡æ®µ
            ns_counter[ns] += 1
    summary["TopNamespaces"] = ns_counter.most_common(8)

    return summary
# ---------- ä¸»æµç¨‹ ----------
g = Graph()
loaded_files = load_with_imports(g, AggregatedUrl, uri_map)
stats = summarize_graph(g)

tmp_xml = tempfile.mkstemp(suffix=".owl")[1]
g.serialize(tmp_xml, format="xml")

# ---------- è¾“å‡ºæ‰“å° ----------
print("\n=== âœ… å·²åŠ è½½æ–‡ä»¶ ===")
for f in loaded_files:
    print("  â€¢", f)

print("\n=== ğŸ“Š å›¾è°±ç»Ÿè®¡ ===")
for k, v in stats.items():
    if k != "TopNamespaces":
        print(f"{k:20s}: {v}")

print("\n=== ğŸ—‚ï¸  å‘½åç©ºé—´å‡ºç°é¢‘æ¬¡ï¼ˆTop 8ï¼‰ ===")
for ns, cnt in stats["TopNamespaces"]:
    print(f"{cnt:>6} triples  {ns}")
````
